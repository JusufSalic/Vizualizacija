# Procjena vrijednosti broja $\pi$



## Monte-Carlo tehnike
Monte Carlo je mali gradić na francuskoj rivijeri poznat po svome kasinu. Zbog slave u oblasti igara na sreću i slučajnih događaja, niz tehnika koje se odnose na slučajne (radnom) metode u riješavanju kompjutacionih problema nazvan je po ovom gradiću. Naprimjer, ako želimo izračunati površinu nekog lika neobičnog/nepravilnog oblika, Monte-Carlo pristup bi bio sljedeći: Ovu površinu opkolimo površinom poznatog oblika (npr. kvadratom ili krugom), a zatim generiramo veliki broj slučajnih brojeva koji leže unutar poznate površine. Površina nepoznatog lika bi tada bila jednaka dijelu tačaka koje leže unutar nepoznate površine pomnožene sa površinim poznatog lika. 

Da bi ove metode funkcionisale, potrebno je da generirani slučajni brojevi zadovoljavaju tri kriterija:
1. Moraju biti ravnomjerno distribuirani. Ako su pak brojevi raspoređeni oko neke normalne krive, onda bi na imali više tačaka oko sredine naše slike i rezultat proračuna bi bio loš.
2. Brojevi moraju biti međusobno nepovezani (ne smije biti korelacije između njh).Naprimjer, kada bi brojevi na slici bili povezani tako da male vrijednosti uvijek prati veliki broj i obrnuto, doili bi sliku  na kojoj bi tačke težile krajnje desno i krajnje lijevo. Rezultat proračuna bi ponovo bio loš. 
3. Potrebno je generirati veliki broj slučajnih brojeva. Greška se ponaša prema zakonu $1/\sqrt{N}$, tako da bi dobili vrijednost greške bila na tri cifre iza decimalnog zareza, trebalo bi nam oko milion parova genirarnih brojeva

Ulazni podaci:

 - N - ukupan broj pokušaja

 - n - broj uspješnih pokušaja    
  
Izlazni podaci: 

- vrijednost broja $\pi$



```python
import numpy as np
N=10000
n=0
x=[]
y=[]
for i in range(N):
    x.append(np.random.uniform())
    y.append(np.random.uniform())
    r=np.sqrt(x[i]**2 + y[i]**2)
    if r <= 1:
        n+=1
PI=4* n/N
print(PI)
xk=np.arange(0, 1, 0.01)
yk=np.sqrt(1-xk**2)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x, y, s=1)
plt.plot(xk, yk, c="r")
plt.xlim(-0.1,1.3)
plt.show()
```

```python
# Definisati cemo kod u obliku funkcije
def PIMC(N):
    N
    n=0
    x=[]
    y=[]
    for i in range(N):
        x.append(np.random.uniform())
        y.append(np.random.uniform())
        r=np.sqrt(x[i]**2 + y[i]**2)
        if r <= 1:
            n+=1
    PI=4* n/N
    return PI, n, x, y

    
xk=np.arange(0, 1, 0.01)
yk=np.sqrt(1-xk**2)
```

```python
PI_1 , n_1, x_1, y_1 =PIMC(10)
PI_2 , n_2, x_2, y_2 =PIMC(1000)
PI_3 , n_3, x_3, y_3 =PIMC(10000)
PI_4 , n_4, x_4, y_4 =PIMC(100000)
```

```python
import matplotlib.pyplot as plt
#plt.style()
plt.figure(figsize=(16,12))
fig1=plt.figure(figsize=(16,12))
fig1.suptitle('Određivanje vrijednoti broja $\pi$ Monte Carlo metodom') #Naslov slike

plt.subplot(2,2,1)      # brojevi (2,2,1) oznacavaju da se radi o multiplotu dimenzija 2x2, i da crtamo u 1. subplotu
plt.scatter(x_1, y_1, s=0.7, label=PI)
plt.plot(xk, yk)
plt.legend(loc=10)


plt.subplot(2,2,2)
plt.scatter(x_2, y_2, s=0.2, label="N=1k")
plt.plot(xk, yk)
plt.legend(loc=10)


plt.subplot(2,2,3)
plt.scatter(x_3, y_3, s=0.2,  label="N=100k")
plt.plot(xk, yk)
plt.legend(loc=10)


plt.subplot(2,2,4)
plt.scatter(x_4, y_4, s=0.2, label="N=1M")
plt.plot(xk, yk)
plt.legend(loc=10)


plt.show()
# 
```
